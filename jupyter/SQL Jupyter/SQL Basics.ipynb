Show the **sum** of prices of NOMBREARTÍCULO, grouped by SECCIÓN, ordered by it sum of price (SUMAPRECIO)

```python
%%sql
select NOMBREARTÍCULO, SECCIÓN, SUM(PRECIO) AS SUMAPRECIO FROM PRODUCTOS GROUP BY SECCIÓN ORDER BY SUMAPRECIO DESC
```

Show the **average** of prices of NOMBREARTÍCULO, grouped by SECCIÓN, ordered by it sum of price (SUMAPRECIO)

```python
%%sql
select NOMBREARTÍCULO, SECCIÓN, AVG(PRECIO) AS SUMAPRECIO FROM PRODUCTOS GROUP BY SECCIÓN ORDER BY SUMAPRECIO DESC
```

**Rounding** decimals:

```python
%%sql
select NOMBREARTÍCULO, SECCIÓN, ROUND(SUM(PRECIO),2)AS SUMAPRECIO FROM PRODUCTOS GROUP BY SECCIÓN ORDER BY SUMAPRECIO DESC
```

```python
%%sql
select NOMBREARTÍCULO, SECCIÓN, ROUND(SUM(PRECIO),2)AS SUMAPRECIO, PRECIO*0.21 AS TAX FROM PRODUCTOS GROUP BY SECCIÓN ORDER BY SUMAPRECIO DESC
```

Dates:

```python
%%sql
select NOMBREARTÍCULO, SECCIÓN, PRECIO, FECHA, NOW() AS DIA_DE_HOY FROM PRODUCTOS WHERE SECCIÓN='DEPORTES'
```

```python
%%sql
select NOMBREARTÍCULO, SECCIÓN, PRECIO, FECHA, NOW() AS DIA_DE_HOY, DATEDIFF(NOW(),FECHA) AS DIFERENCIADIAS FROM PRODUCTOS WHERE SECCIÓN='DEPORTES'
```

```python
%%sql
select NOMBREARTÍCULO, SECCIÓN, PRECIO, DATE_FORMAT(FECHA, '%D-%M-%Y') AS FECHA, DATE_FORMAT(NOW(), '%D-%M-%Y') AS DIA_DE_HOY, DATEDIFF(NOW(),FECHA) AS DIFERENCIADIAS FROM PRODUCTOS WHERE SECCIÓN='DEPORTES'
```

## Multiple table queries

- External Union
    - Union
    - Union All
    - Except (not accepted by MySQL)
    - Intesect (not accepted by MySQL)
    - Minus (not accepted by MySQL)
- Internal Union
    - Inner Join
    - Left Join
    - Right Join

### Union

- Both tables must have same number of fields
- Fields must have compatible types
- Fields can have different fieldnames

*Note: Union concatenates second table under the first one, and takes fieldnames from table1

Basically, we concatenate two (or more) SELECT queries by the word **union**

Selecting all columns:

```python
%%sql

SELECT * FROM PRODUCTOS WHERE SECCIÓN='DEPORTES' UNION SELECT * FROM PRODUCTOSNUEVOS WHERE SECCIÓN='DEPORTES DE RIESGO'
```

Selecting some columns:

```python
%%sql
SELECT CÓDIGOARTÍCULO, SECCIÓN FROM PRODUCTOS WHERE SECCIÓN='DEPORTES' UNION SELECT CÓDIGOARTÍCULO, SECCIÓN FROM PRODUCTOSNUEVOS WHERE SECCIÓN='DEPORTES DE RIESGO'
```

```python
%%sql
SELECT CÓDIGOARTÍCULO, SECCIÓN FROM PRODUCTOS WHERE SECCIÓN='DEPORTES' UNION SELECT CÓDIGOARTÍCULO, SECCIÓN FROM PRODUCTOSNUEVOS WHERE SECCIÓN='DEPORTES DE RIESGO'
```

### UNION ALL 
This is basically the same as **union**, but it shows any row, even though it is repeated between tables

### Internal Union

- Inner Join: all common rows
- Left Join: Left table rows + common rows
- Right Join: common rows + Right table rows

In order to continue testing joins, we have to be sure that CLIENTES and PEDIDOS tables are connected by indexes and foreign keys:

- CLIENTES: 
    - Primary Key: CÓDIGOCLIENTE (UNIQUE)
- PEDIDOS:
    - Primary Key: NÚMERO DE PEDIDO (UNIQUE)
    - Index: CÓDIGO CLIENTE

Showing clients that already ordered something:

```python
%%sql

SELECT CLIENTES.CÓDIGOCLIENTE, POBLACIÓN, DIRECCIÓN, `NÚMERO DE PEDIDO`, 
PEDIDOS.`CÓDIGO CLIENTE`, `FORMA DE PAGO` 
FROM CLIENTESPILDORAS.CLIENTES INNER JOIN CLIENTESPILDORAS.PEDIDOS 
ON CLIENTES.CÓDIGOCLIENTE=PEDIDOS.`CÓDIGO CLIENTE` 
WHERE POBLACIÓN='MADRID';
```

Now we change INNER JOIN to **LEFT JOIN**, so we will have both customers that have placed at least one order, and customers who have not ordered anything:

```python
%%sql

SELECT CLIENTES.CÓDIGOCLIENTE, POBLACIÓN, DIRECCIÓN, `NÚMERO DE PEDIDO`, 
PEDIDOS.`CÓDIGO CLIENTE`, `FORMA DE PAGO` 
FROM CLIENTESPILDORAS.CLIENTES left JOIN CLIENTESPILDORAS.PEDIDOS 
ON CLIENTES.CÓDIGOCLIENTE=PEDIDOS.`CÓDIGO CLIENTE` 
WHERE POBLACIÓN='MADRID';
```

Customers who have **"none"** CÓDIGO CLIENTE cell, are those who have not placed an order yet;


If we want to check those clients those customers who have not yet placed an order, we add a **is null** clause:


```python
%%sql

SELECT CLIENTES.CÓDIGOCLIENTE, POBLACIÓN, DIRECCIÓN, `NÚMERO DE PEDIDO`, 
PEDIDOS.`CÓDIGO CLIENTE`, `FORMA DE PAGO` 
FROM CLIENTESPILDORAS.CLIENTES left JOIN CLIENTESPILDORAS.PEDIDOS 
ON CLIENTES.CÓDIGOCLIENTE=PEDIDOS.`CÓDIGO CLIENTE` 
WHERE POBLACIÓN='MADRID'
AND PEDIDOS.`CÓDIGO CLIENTE` is null;
```

Using **ALL** operator:
- The ALL operator returns true if all of the subquery values meet the condition.

Show all products whose price is higher than all 'CERAMICA' products

```python
%%sql

select PRECIO from PRODUCTOS where SECCIÓN='CERÁMICA'
```

```python
%%sql

select * from PRODUCTOS 
where PRECIO > ALL (select PRECIO from PRODUCTOS where SECCIÓN='CERÁMICA')
```

Using **ANY** operator:
- The ANY operator returns true if any of the subquery values meet the condition.

Show all products whose price is higher than **any** of 'CERAMICA' products

```python
%%sql

select * from PRODUCTOS 
where PRECIO > ANY (select PRECIO from PRODUCTOS where SECCIÓN='CERÁMICA')
```

*Note: in this case, all is like comparing to the highest price for one given category, and any is like comparing to the lower price for one given category.

more **all** and **any** examples:

```python
%%sql

select * from PRODUCTOS where PRECIO > all (select PRECIO from PRODUCTOS where SECCIÓN='JUGUETERÍA')
```

```python
%%sql

select * from PRODUCTOS where PRECIO > any (select PRECIO from PRODUCTOS where SECCIÓN='JUGUETERÍA')
```

**In** operator:

Show all items that have been ordered in quantities higher than 20 units

```python
%%sql

select `CÓDIGO ARTÍCULO` from CLIENTESPILDORAS.PRODUCTOSPEDIDOS where UNIDADES>20
```

```python
%%sql

select NOMBREARTÍCULO, PRECIO 
from CLIENTESPILDORAS.PRODUCTOS 
where CÓDIGOARTÍCULO in (select `CÓDIGO ARTÍCULO` from CLIENTESPILDORAS.PRODUCTOSPEDIDOS where UNIDADES>20)
```

Same query but with an **inner join**:

```python
%%sql

select NOMBREARTÍCULO, PRECIO 
from PRODUCTOS inner join PRODUCTOSPEDIDOS 
on PRODUCTOS.CÓDIGOARTÍCULO=PRODUCTOSPEDIDOS.`CÓDIGO ARTÍCULO` 
where UNIDADES>20
```
